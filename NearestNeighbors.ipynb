```python
# credit to carlos theran and yohn jairo
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import NearestNeighbors
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
#loading Iris Data sets,
# This data sets consists of 3 different types of irisesâ€™ (Setosa, Versicolour, and Virginica) 
# Stored in a 150x4 numpy.ndarray
# The rows being the samples and the columns being: Sepal Length, Sepal Width, Petal Length and Petal Width.
#https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris
data = load_iris()
```

```python
# print the class of the samples 10, 25, and 50.
data.target[[10,25,50]]
```

```python
# print the label or class name of your datasets.
list(data.target_names)
```

```python
# import some data to play with
X = data.data[:, :2]  # we only take the first two features.
y = data.target
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
plt.figure(2, figsize=(8, 6))
plt.clf()
# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python

X = data.data[:, 2:4]  # we only take the last two features.
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
plt.figure(2, figsize=(8, 6))
plt.clf()
# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python
# Dividing the data into training and testing create data input to feed the model
#https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.NearestNeighbors.html#sklearn.neighbors.NearestNeighbors.kneighbors
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
nbrs = NearestNeighbors(n_neighbors=3, algorithm='ball_tree').fit(X)
```

```python
# reshape the input in order to match with the training input data and find the distances and indices from each training points.
distances, indices = nbrs.kneighbors(X_test.reshape(-1,4), return_distance=True, )
```

```python
# Print the distances 
distances.shape
```

```python
#now Lets work with KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)
```

```python
# used the trained model to predict the label or classes of the test data.
classification_restuls = neigh.predict(X_test)
```

```python
#Print the results of your prediction
classification_restuls
```

```python
y_test
```

```python
#Find the performance of your model
neigh.predict_proba(X_test)
```

```python


# Plot non-normalized confusion matrix
from sklearn.metrics import plot_confusion_matrix
titles_options = [("Confusion matrix, without normalization", None)]
for title, normalize in titles_options:
    disp = plot_confusion_matrix(neigh, X_test, y_test,
                                 cmap=plt.cm.Blues,
                                 normalize=normalize)
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()

```
